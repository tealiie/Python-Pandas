```python
# Import pandas
import pandas as pd

csv_path = "Resources/purchase_data.csv"

purchase_data_df = pd.read_csv(csv_path)
purchase_data_df.head(15)
```

```python
purchase_data_df["Age"].max()
```

```python
# Print Player Count
player_count = purchase_data_df["SN"].nunique()
total_player_df = pd.DataFrame({"Total Players" : [player_count]})
total_player_df
```

```python
unique_items = purchase_data_df["Item Name"].value_counts()
avg_price = round(purchase_data_df["Price"].mean(),2) 
purchases = purchase_data_df["Purchase ID"].count()
revenue = purchase_data_df["Price"].sum()

purchases_summary_df = pd.DataFrame({"Number of Unique Items": len(unique_items),
                                    "Average Price": [avg_price],
                                    "Number of Purchases": [purchases],
                                    "Total Revenue": [revenue]})

purchases_summary_df["Average Price"] = purchases_summary_df["Average Price"].map("${:.2f}".format)
purchases_summary_df["Total Revenue"] = purchases_summary_df["Total Revenue"].map("${:,.2f}".format)
                        
purchases_summary_df
```

```python
gender_count = purchase_data_df.groupby("Gender")["SN"].nunique()
gender_percentage = gender_count / player_count


gender_summary_df = pd.DataFrame({"Total Count": gender_count,
                                 "Percentage of Players": gender_percentage})

gender_summary_df["Pecentage of Players"] = gender_summary_df["Percentage of Players"].map("{:.2%}".format)

gender_summary_df
```

```python
purchase_count = purchase_data_df.groupby("Gender")["Purchase ID"].count()
total_purchase = purchase_data_df.groupby("Gender")["Price"].sum()
avg_purchase = total_purchase / purchase_count
avg_purchase_perperson = total_purchase / gender_count

gender_purchase_summary_df = pd.DataFrame({"Purchase Count": purchase_count,
                                           "Average Purchase Price": avg_purchase,
                                           "Total Purchase Value": total_purchase,
                                           "Avg Total Purchase per Person": avg_purchase_perperson})

gender_purchase_summary_df["Average Purchase Price"] = gender_purchase_summary_df["Average Purchase Price"].map("${:.2f}".format)
gender_purchase_summary_df["Total Purchase Value"] = gender_purchase_summary_df["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_summary_df["Avg Total Purchase per Person"] = gender_purchase_summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)

gender_purchase_summary_df
```

```python
bins = [0, 9, 14, 19, 24, 29, 34 , 39, 45]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Groups"] = pd.cut(purchase_data_df["Age"], bins, labels = group_names, include_lowest=True)

player_age_count = purchase_data_df.groupby("Age Groups")["SN"].nunique()
percentage_age_player = player_age_count / player_count

age_demo_df = pd.DataFrame ({"Total Count": player_age_count,
                             "Percentage of Players": percentage_age_player})

age_demo_df["Pecentage of Players"] = age_demo_df["Percentage of Players"].map("{:.2%}".format)

age_demo_df
```
